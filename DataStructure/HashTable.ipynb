# Hash Table
---
- (key, value): 키(key)에 데이터(value)를 매핑할 수 있는 데이터 구조
- key를 통해 데이터가 저장된 주소를 바로 알 수 있어, 저장 및 탐색 속도가 매우 빠름
- 해쉬 함수(Hash Function): 데이터를 고정된 길이의 값으로 바꿔주는 함수
- 해쉬 테이블(Hash Table): 키 값의 연산으로 직접 데이터에 접근 가능한 데이터 구조



```python
class HashTable:
    def __init__(self, table_size):
        self.size = table_size
        self.hash_table = [0 for a in range(self.size)]
        
    def getKey(self, data):
        self.key = ord(data[0])
        return self.key
    
    def hashFunction(self, key):
        return key % self.size
    
    def getAddress(self, key):
        myKey = self.getKey(key)
        hash_address = self.hashFunction(myKey)
        return hash_address
    
    def save(self, key, value):
        hash_address = self.getAddress(key)
        self.hash_table[hash_address] = value
    
    def read(self, key):
        hash_address = self.getAddress(key)
        return self.hash_table[hash_address]
    
    def delete(self, key):
        hash_address = self.getAddress(key)
        if self.hash_table[hash_address] != 0:
            self.hash_table[hash_address] = 0
            return True
        else:
            return False
if __name__ == '__main__':
    ht = HashTable(8)
    ht.save('a', 1111)
    ht.save('b', 3333)
    ht.save('c', 5555)
    print(ht.hash_table)
    print(ht.read('c'))
    ht.delete('c')
    print(ht.hash_table)
```

# Hash Collision
 ### Chaining 기법
 ---
 - Open Hashing 기법 중 하나로, 충돌이 발생하면 링크드 리스트를 이용하여 데이터를 뒤에 연결시켜서 저장
 
### Linear Probing 기법
 ---
 - Close Hashing 기법 중 하나로, 충돌한 hash address의 다음 맨 처음 빈 공간에 저장
