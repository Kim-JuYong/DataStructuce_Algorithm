# Tree
---
- Node와 Branch를 이용해서, 사이클을 이루지 않도록 수성한 데이터 구조
- 용어
    - Node: 트리에서 데이터를 저장하는 기본 요소
    - Root Node: 트리 맨 위에 있는 최상위 노드
    - Level: 최상위 노드를 Level 0으로 하였을 때, 하위 Branch로 연결된 노드들의 깊이
    - Parent Node: 상위 Branch로 연결된 노드
    - Child Node: 하위 Branch로 연결된 노드
    - Leaf Node: Child Node가 존재하지 않는 노드
    - Slibing(Brother Node): 동일한 Parent Node를 가지는 노드
    - Depth: 트리에서 Node가 가질 수 있는 최대 Level

# Binary Search Tree
---
- 이진 트리(Binary Tree): 노드가 Left Child Node와 Right Child Node만을 가지는 Tree
- 이진 탐색 트리(Binary Search Tree): Left Child Node는 해당 노드보다 작은 값을, Right Child Node는 해당 노드보다 큰 값을 가지는 Tree
