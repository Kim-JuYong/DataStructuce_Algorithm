# Queue
---
- FIFO(First-In-First-Out): 가장 먼저 넣은 데이터를 가장 먼저 꺼낼 수 있는 데이터 구조
- queue모듈의 Queue클래스를 이용하여 생성
- 데이터 맨 앞에 삽입: q.put(element)
- 첫번째 데이터 반환∙삭제: q.get()


```python
from queue import Queue

q = Queue() #생성자
q.put(1) #데이터를 맨 앞에 삽입
q.put(2)
q.put(3)
print(q.get()) #첫번째 데이터 반환∙삭제
print(q.get())
```

# Deque
---
- 데이터를 양방향에서 추가하고 제거 가능한 데이터 구조
- 데이터 맨 앞에 삽입: deq.appendleft(element)
- 데이터 맨 마지막에 삽입: deq.append(element) 
- 첫번째 데이터 반환∙제거: deq.popleft()
- 마지막 데이터 반환∙제거: deq.pop()

```python
from collections import deque
deq = deque([1, 2, 3]) #생성
deq.append(4) #데이터 맨 끝 삽입
deq.appendleft(0) #데이터 맨 처음 삽입
print(deq)
print(deq, deq.pop()) #데이터 맨 끝 반환∙제거
print(deq, deq.popleft()) #데이터 맨 처음 반환∙제거
```

# Priority Queue
---
- 우선 순위가 낮은 데이터를 가장 먼저 꺼낼 수 있는 데이터 구조
- 데이터 삽입: pq.put((우선순위, element))
- 데이터 반환∙제거: pq.get()

```python
import queue

pq = queue.PriorityQueue()
pq.put((10, "Hello")) #(우선 순위, 데이터) 의 튜플 구조로 삽입
pq.put((14, "World"))
pq.put((13, "-"))
print(pq.get()) #우선 순위가 가장 낮은 데이터 반환∙제거
print(pq.get())
```
