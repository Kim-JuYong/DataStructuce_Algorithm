# Queue
---
- FIFO(First-In-First-Out):           
- queue Queue  
-    : q.put(element)
-   : q.get()


```python
from queue import Queue

q = Queue() #
q.put(1) #   
q.put(2)
q.put(3)
print(q.get()) #  
print(q.get())
```

# Deque
---
-       
-    : deq.appendleft(element)
-    : deq.append(element) 
-   : deq.popleft()
-   : deq.pop()

```python
from collections import deque
deq = deque([1, 2, 3]) #
deq.append(4) #   
deq.appendleft(0) #   
print(deq)
print(deq, deq.pop()) #   
print(deq, deq.popleft()) #   
```

# Priority Queue
---
-           
-  : pq.put((, element))
-  : pq.get()

```python
import queue

pq = queue.PriorityQueue()
pq.put((10, "Hello")) #( , )    
pq.put((14, "World"))
pq.put((13, "-"))
print(pq.get()) #     
print(pq.get())
```
