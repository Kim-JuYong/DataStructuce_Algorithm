# Linked List
---
- 떨어진 곳에 존재하는 데이터를 포인터로 연결해서 관리하는 데이터 구조 -> 파이썬은 list 자료형에 linked list가 포함됨
- 노드(Node): 데이터의 저장 단위 (데이터값, 포인터)로 구성됨
- 포인터(pointer): 각 노드 안에서, 다음이나 이전의 노드 연결 정보를 가지고 있는 공간
