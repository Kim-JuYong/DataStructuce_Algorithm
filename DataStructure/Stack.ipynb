# Stack
---
- LIFO(Last-In-First-Out): 가장 나중에 넣은 데이터를 가장 먼저 빼낼 수 있는 데이터 구조 -> 파이썬에서는 리스트로 대체
- 마지막에 데이터 삽입: s.append(element) 
- 마지막 데이터 제거: s.pop()
- 마지막 데이터 가져오기: s[-1]


```python
stack = []
stack.append(1)
stack.append(2)
stack.append(3)
print(stack)

print(stack[-1])

print(stack, stack.pop())
print(stack, stack.pop())
```
